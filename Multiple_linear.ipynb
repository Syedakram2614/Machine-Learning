```python
#project use case: multiple linear regression
#y=mx1+mx2+......+mxn+c
#y=beta1x1+beta2x2+......+beta0+error
#independent variable:x1,x2,.....,xn
#dependent variable:y

#------------------------------------------------------------------------------------
#usecase: we are analysing the profit of a 50 startup company located in different cities using mulitples linear regression model.
#stpe1: import all the required libraries
#step2: EDA1 process
#step3: feature extraction
#step4: one hot encoding(NLP)
#step5: EDA2 process- training and testing the model
#step6: model building
#step7: prediciting the values
#step8: evalutaion of the model
#------------------------------------------------------------------------------------


import pandas as pd
df=pd.read_csv("50_Startups_Sample.csv")
```

```python
#feature extraction
x=df.iloc[:,:-1]
```

```python
x
```

```python
 y=df["Profit"]
```

```python
y
```

```python
#one hot encoding - it is a nlp technique used in macine learning and data preprocessing technique to convert categirca data to numerical data
#nlp - it is a field of artificial intelligence that enables computer to understand interpret and generate human language.
#States=pd.get_dummies(x['State'],drop_first=True)
#x=x.drop('State',axis=1)
#x=pd.concat([x,States],axis=1)
#axis=1 means drop one column
```

```python
states = pd.get_dummies(df['State'], drop_first=True)
x = df.drop(['State', 'Profit'], axis=1)
x = pd.concat([x, states], axis=1)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
```

```python
LR.fit(x_train,y_train)
```

```python
y_pred=LR.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
```
